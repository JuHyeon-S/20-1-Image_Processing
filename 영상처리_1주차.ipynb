```
import cv2                            #OpenCV 라이브러리
import matplotlib.pyplot as plt       #자료를 차트나 플롯(plot)으로 시각화하는 패키지
%matplotlib inline

img = cv2.imread('./input.jpg')       #이미지 파일 읽기

plt.imshow(img[:,:,::-1])             #RGB로 변형
plt.show()                            #그림을 출력하는 함수
```
