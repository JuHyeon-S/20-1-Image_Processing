```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./carrot.bmp')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

fig = plt.figure(figsize=(15, 10))

img_gray_add = img_gray + 70
img_gray_sub = img_gray - 70
img_gray_mul = img_gray * 1.5
img_gray_div = img_gray / 1.5

img_gray_mul = img_gray_mul.astype(np.int)
img_gray_div = img_gray_div.astype(np.int)

ax1 = fig.add_subplot(221)
ax1.set_title('carrot (Addtion)')
ax1.imshow(img_gray_add, cmap='gray')

ax1 = fig.add_subplot(222)
ax1.set_title('carrot (SUbtraction)')
ax1.imshow(img_gray_sub, cmap='gray')

ax1 = fig.add_subplot(223)
ax1.set_title('carrot (Multiplication)')
ax1.imshow(img_gray_mul, cmap='gray')

ax1 = fig.add_subplot(224)
ax1.set_title('carrot (Division)')
ax1.imshow(img_gray_div, cmap='gray')

plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./carrot.bmp')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

fig = plt.figure(figsize=(15, 10))

img_gray_add = cv2.add(img_gray, 70)
img_gray_sub = cv2.subtract(img_gray, 70)
img_gray_mul = cv2.multiply(img_gray, 1.5)
img_gray_div = cv2.divide(img_gray, 1.5)

img_gray_mul = img_gray_mul.astype(np.int)
img_gray_div = img_gray_div.astype(np.int)

ax1 = fig.add_subplot(221)
ax1.set_title('carrot (Addtion)')
ax1.imshow(img_gray_add, cmap='gray')

ax1 = fig.add_subplot(222)
ax1.set_title('carrot (SUbtraction)')
ax1.imshow(img_gray_sub, cmap='gray')

ax1 = fig.add_subplot(223)
ax1.set_title('carrot (Multiplication)')
ax1.imshow(img_gray_mul, cmap='gray')

ax1 = fig.add_subplot(224)
ax1.set_title('carrot (Division)')
ax1.imshow(img_gray_div, cmap='gray')

plt.show()
```

lichtenstein,bmp

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Inimg = cv2.imread('./Lichtenstein.bmp')
Inimg = cv2.cvtColor(Inimg, cv2.COLOR_BGR2RGB)

table = np.zeros(256).astype(np.int)

for i in range(256):
  temp = int(i * 0.5)
  table[i] = 255 if temp > 255 else temp

OutImg1 = cv2.LUT(Inimg, table)

for i in range(256):
  temp = int(i * 1.5)
  table[i] = 255 if temp > 255 else temp

OutImg2 = cv2.LUT(Inimg, table)

for i in range(256):
  temp = int(i * 2.0)
  table[i] = 255 if temp > 255 else temp

OutImg3 = cv2.LUT(Inimg, table)

fig = plt.figure(figsize=(15, 10))

ax1 = fig.add_subplot(231)
ax1.set_title('Lookup Table(0.5)')
ax1.imshow(OutImg1)

ax1 = fig.add_subplot(232)
ax1.set_title('Lookup Table(1.5)')
ax1.imshow(OutImg2)

ax1 = fig.add_subplot(233)
ax1.set_title('Lookup Table(2.0)')
ax1.imshow(OutImg3)

plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

Inimg = cv2.imread('./Lichtenstein.bmp')
Inimg = cv2.cvtColor(Inimg, cv2.COLOR_BGR2RGB)

level1 = 3
level2 = 6
table = np.zeros(256).astype(np.int)
step = 256 / level1
inc = 255 / (level1 - 1)

for i in range(256):
  table[i] = int(i /step) * inc

OutImg1 = cv2.LUT(Inimg, table)

step = 256 / level2
inc = 255 / (level2 - 1)

for i in range(256):
  table[i] = int(i /step) * inc

OutImg2 = cv2.LUT(Inimg, table)


fig = plt.figure(figsize=(15, 10))

ax1 = fig.add_subplot(221)
ax1.set_title('Posterizing Table(3)')
ax1.imshow(OutImg1)


ax1 = fig.add_subplot(222)
ax1.set_title('Posterizing Table(6)')
ax1.imshow(OutImg2)

plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./carrot.bmp')

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

channels = cv2.split(img_rgb)
colors = ['red', 'green', 'blue']

plt.title('carrot (3channel RGB)')
for (ch, co) in zip(channels, colors):
  hist = cv2.calcHist([ch], [0], None, [256], [0, 256])
  plt.plot(hist, color=co)
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./carrot.bmp')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_normarlize = cv2.normalize(img_gray, None, 0, 255, cv2.NORM_MINMAX)
img_equalize = cv2.equalizeHist(img_gray)

hist1 = cv2.calcHist([img_normarlize], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([img_equalize], [0], None, [256], [0, 256])

title = ['carrot (normalize Historgram)', 'carrot (equalize Historgram)']
histogram = [hist1, hist2]

fig = plt.figure(figsize=(10, 5))

idx = 1
for t, h in zip(title, histogram):
  plt.subplot(1,2,idx)
  plt.title(t)
  plt.plot(h)
  idx += 1
plt.show()
```

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./carrot.bmp')

img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

hist1 = cv2.calcHist([img_gray], [0], None, [256], [0, 256])

fig = plt.figure(figsize=(10, 5))

plt.subplot(1,2,1)
plt.title('carrot (1channel GRAY)')
plt.plot(hist1)

channels = cv2.split(img_rgb)
colors = ['red', 'green', 'blue']

plt.subplot(1,2,2)
plt.title('carrot (3channel RGB)')
for (ch, co) in zip(channels, colors):
  hist2 = cv2.calcHist([ch], [0], None, [256], [0, 256])
  plt.plot(hist2, color=co)
plt.show()
```
